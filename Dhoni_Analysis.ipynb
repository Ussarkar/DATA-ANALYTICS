```python
#Step 3 :Comparision between other Players of TATA IPL 2023 with similar record.
```

```python
AP1= pd.DataFrame(df.groupby('striker')['runs_off_bat'].sum()).reset_index()
```

```python
AP2= pd.DataFrame(df.groupby('striker')['ball'].count()).reset_index()
```

```python
AP2= AP1.merge(AP2,on = 'striker', how='left')
AP2= AP2[AP2.runs_off_bat >= 300]
AP2= AP2[AP2.ball <= 270]
AP2.head()
```

```python
AP2['strike_rate']= AP2['runs_off_bat']/AP2['ball']*100
```

```python
AP2.reset_index(inplace=True, drop=True)
AP2.head(4)
```

```python
AP2.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = False).plot(kind ='barh')
plt.xlabel('Runs scored')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,9))
plt.scatter(AP2.strike_rate, AP2.runs_off_bat)

for i in range(len(AP2)):
     plt.text(AP2['strike_rate'][i] + 1, AP2['runs_off_bat'][i] - 1, AP2['striker'][i] )

plt.title('Batsmen with more than 300+ runs scored with less than 270 balls(Finishers) from IPL 2021-2023', fontsize=26 )
plt.xlabel('Strike rate', fontsize=16)
plt.ylabel('Runs scored', fontsize=16)
plt.axvline(130, ls='--', color='black')
plt.axhline(270, ls='--', color='black')
plt.show()
```
